```python
import json
import pyperclip
import requests
from pprint import pprint
import pandas as pd
import time

def req(url):
    res = requests.get(url).json()
    pyperclip.copy(json.dumps(res))
    #pprint(res)
    return res
```

```python
def fetch_static(file_path, accounts):
    def filter_data(obj):
        return {k: v for k, v in obj.items() if isinstance(v, (int, float)) and not isinstance(v, (bool))}

    rows = pd.DataFrame()
    for account in accounts:
        t = req('https://alpari.com/api/ru/pamm/{}/'.format(account))
        time.sleep(.1)

        pamm = filter_data(t['data']['pamm'])
        activities = filter_data(t['data']['pamm']['activities'])
        row = pd.DataFrame.from_dict([{**pamm, **activities, 'accountId': account}])

        rows = pd.concat([rows, row], axis=0, ignore_index=True)
    rows = rows.set_index('accountId', drop=True)
    rows.to_csv(file_path, index=True)
```
